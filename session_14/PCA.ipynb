```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib as mpl
import matplotlib.pyplot as plt

breastData = load_breast_cancer()
```

```python
breastData.data
```

```python
breastData.target
```

```python
labels = np.reshape(breastData.target, (569,1))
```

```python
breast_data = np.concatenate([breastData.data, labels], axis=1)
```

```python
breastDS = pd.DataFrame(breast_data)
```

```python
breastFeatureLabel = np.append(breastData.feature_names,"label")
```

```python
breastDS.columns = breastFeatureLabel
```

```python
breastDS.head()
```

```python
breastDS['label_name']  = np.where(breastDS["label"] == 0 , "Benign", "Maligant")
```

```python
breastDS.tail(20)
```

```python
x= breastDS.loc[:, breastData.feature_names].values
x= StandardScaler().fit_transform(x)
```

```python
column_name = ["feature"+str(i) for i in range(x.shape[1])]
```

```python
normalize_breast_df = pd.DataFrame(x,columns= column_name)
```

```python
normalize_breast_df.head()
```

```python
pca_model = PCA(n_components=2)
principal_com_breash = pca_model.fit_transform(x)
```

```python
principal_com_breash
```

```python
pca_model.explained_variance_ratio_
```

```python
pca_breast_df = pd.DataFrame(principal_com_breash,columns= ["principal 1", "principal 2"])
```

```python
pca_breast_df
```

```python
plt.figure()
plt.figure(figsize=(15,15))
plt.xlabel("principal Comonent -1 ")
plt.ylabel("principal Comonent -2 ")
plt.title("Principal componenr breash cancer  ")

targets =["Benign", "Maligant"]
colors = ["r","g"]

for target, color in zip(targets,colors ):
    idxToKeep = breastDS["label_name"] == target
    plt.scatter(pca_breast_df.loc[idxToKeep, "principal 1"],
                pca_breast_df.loc[idxToKeep, "principal 2"],
                c=color, s= 40)

plt.legend(targets)
```
