```python
movieDB = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movieDB.head()
```

```python
movieDB.describe()
```

```python
movieDB.info()
```

```python
# filter film dengan duration <= 100 , star_rating > 7, urutkan dari rating tertinggi

veryGoodMovie = movieDB[(movieDB.star_rating > 7 )& (movieDB.duration <=100) ]
veryGoodMovie.sort_values(by=["star_rating"], ascending=False)
```

```python
import re
movieDB[movieDB.title.str.contains("^sp", flags=re.IGNORECASE)]groupby(['genre'])
```

```python
groupedMovie = movieDB.groupby(['genre'])["title"]
groupedMovie.describe()
```

```python
summary = movieDB.pivot_table(index=["genre"], 
                              values = ["duration", "star_rating"], 
                              aggfunc={"duration":"mean", 
                                       "star_rating":"mean"})
```

```python
summary
```

```python
summary.plot.bar(subplots=True)
```

```python
summary.to_csv("./moviesum.csv")
```
